```python
def convert_str_to_int(i):
    thousands = i.split(",")
    if len(thousands) == 1:
        return int(thousands[0])
    return int(thousands[0]) * 1000 + int(thousands[1])
```

```python
from urllib.request import urlopen

from bs4 import BeautifulSoup

for i in df[df["total_issue_count"] == 0].index:
    
    url = df.loc[i, "url"] + "/issues"
    
    if url == "https://github.com/notAI-tech/NudeNet/issues":
        continue

    page = urlopen(url)

    html = page.read().decode("utf-8")

    soup = BeautifulSoup(html, "html.parser")
    
    open_issue = ""
    closed_issue = ""
    buttons = soup.find_all("a", {"class": "btn-link"})

    
    for idx, j in enumerate(buttons[:2]):
        if idx == 0:
            open_issue = j.text.strip().split(" ")[0]
        else:
            closed_issue = j.text.strip().split(" ")[0]
            
    
            
    
    
    try:
        total_issue = convert_str_to_int(open_issue) + convert_str_to_int(closed_issue)
        print(f"{url}: open issue: {open_issue}, closed issue:{closed_issue}, total issue: {total_issue}")
        df.loc[i, "total_issue_count"] = total_issue
    except Exception as e:
        continue
```

## Feature Engineering

### Add how many years the project has been developed

```python
df["created_at"] = pd.to_datetime(df["created_at"])
df["updated_at"] = pd.to_datetime(df["updated_at"])
df["years"] = (df["updated_at"] - df["created_at"]) / np.timedelta64(1, 'Y')
df["years"] = df["years"].astype(np.int32)
```

### Find the closed issue count, add closed issue rate (closed / all)

```python
df["closed_issue"] = df["total_issue_count"] - df["open_issues"]
df["issue_close_rate"] = df["closed_issue"] / df["total_issue_count"]
df["issue_close_rate"] = df["issue_close_rate"].fillna(0)
df[["closed_issue", "issue_close_rate", "total_issue_count"]]
```

## Preparing Train and Test Data
We will split train to test as 80%, 20&

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=50, test_size=0.2)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

## Apply Baseline Machine Learning Model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
for i in range(1, 21):
    neigh = KNeighborsClassifier(n_neighbors=i)
    neigh.fit(X_train, y_train)
    preds = neigh.predict(X_test)
    print(f"for neighbour: {i} : accuracy: {accuracy_score(y_test, preds)}")
```

# Different Machine Learning Models

- support vector machine
- gaussian naive bayes

```python
from sklearn import svm
from sklearn.naive_bayes import GaussianNB
from sklearn import tree

models = {
    "SVC": svm.SVC(),
    "Naive Bayes": GaussianNB(),
    "Decision Tree": tree.DecisionTreeClassifier(),
}

for name, model in models.items():

    model.fit(X_train, y_train)

    preds = model.predict(X_test)

    print(f"Accuracy score for {name} : {accuracy_score(y_test, preds)}")
```

```python
model = svm.SVC()

model.fit(X_train, y_train)

preds = model.predict(X_test)

print(f"Accuracy score for SVC: {accuracy_score(y_test, preds)}")
```

```python
model = GaussianNB()

model.fit(X_train, y_train)

preds = model.predict(X_test)

print(f"Accuracy score for Naive Bayes: {accuracy_score(y_test, preds)}")
```

```python
model = tree.DecisionTreeClassifier()

model.fit(X_train, y_train)

preds = model.predict(X_test)

print(f"Accuracy score for Decision Tree: {accuracy_score(y_test, preds)}")
```
